```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("201809-fordgobike-tripdata.csv")
df=df.dropna()
df.head()
```

```python
df["duration_hrs"]=df["duration_sec"]/360000.
df["age"]=2019-df["member_birth_year"]
df.head()
```

```python
df["start_day"]=pd.to_datetime(df["start_time"], errors='ignore')
df["start_day"]= df['start_day'].dt.floor("d")
df.head()
```

```python
df=pd.get_dummies(columns=["member_gender","user_type"],data=df)
df.head()
```

```python
aggregations = {
    'duration_hrs':'mean',
    "age" :"mean",
    "member_gender_Female":"sum",
    "member_gender_Male":"sum",
    "member_gender_Other":"sum",
    "user_type_Customer":"sum",
    "user_type_Subscriber":"sum",  
}
day=df.groupby("start_day").agg(aggregations)
day.head()
```

```python
day["total_riders"]=day["user_type_Customer"]+day["user_type_Subscriber"]
```

```python
print (len(day["total_riders"]))
nextDay=list(day["total_riders"])
nextDay.pop(0)#Don't need this value anymore
nextDay.append(0.0)#Add a zero to the next one as a test
print (len(nextDay))
print (nextDay)
```

```python
from sklearn.preprocessing import RobustScaler
scaledDF = RobustScaler().fit_transform(day)
scaledDF = pd.DataFrame(data=scaledDF, columns = ["scaled_"+str(x) for x in day.columns])
```

```python
scaledDF["nextDay"]=nextDay
scaledDF=scaledDF[:29]
scaledDF.head(30)
```

```python
import seaborn as sbn
g = sbn.pairplot(scaledDF);
```

```python
import seaborn as sns
corr = scaledDF.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

```python
import pymc3 as pm
print('Running on PyMC3 v{}'.format(pm.__version__))
```

```python
basic_model = pm.Model()

with basic_model:

    # Priors for unknown model parameters
    alpha = pm.Normal('alpha', mu=0, sd=10)
    beta = pm.Normal('beta', mu=0, sd=10, shape=1)
    sigma = pm.HalfNormal('sigma', sd=1)

    # Expected value of outcome
    mu = alpha + beta[0]*scaledDF['scaled_age']

    # Likelihood (sampling distribution) of observations
    Y_obs = pm.Normal('Y_obs', mu=mu, sd=sigma, observed=scaledDF["nextDay"])
```

```python
map_estimate = pm.find_MAP(model=basic_model)
map_estimate
```

```python
with basic_model:
    step = pm.NUTS()
    trace = pm.sample(10000, step=step)
```

```python
pm.traceplot(trace);
```

```python
big_model = pm.Model()

with big_model:

    # Priors for unknown model parameters
    alpha = pm.Normal('alpha', mu=0, sd=10)
    beta = pm.Normal('beta', mu=0, sd=10, shape=7)
    sigma = pm.HalfNormal('sigma', sd=1)

    # Expected value of outcome
    """
    ['scaled_duration_hrs', 'scaled_age', 'scaled_member_gender_Female',
       'scaled_member_gender_Male', 'scaled_member_gender_Other',
       'scaled_user_type_Customer', 'scaled_user_type_Subscriber',
       'scaled_start_station_id', 'scaled_total_riders']
    """    
    mu = alpha + beta[0]*scaledDF['scaled_duration_hrs'] +  beta[1]*scaledDF['scaled_age']  \
            +  beta[1]*scaledDF['scaled_member_gender_Female']\
            +  beta[2]*scaledDF['scaled_member_gender_Male']\
            +  beta[3]*scaledDF['scaled_member_gender_Other']\
            +  beta[4]*scaledDF['scaled_user_type_Customer']\
            +  beta[5]*scaledDF['scaled_user_type_Subscriber']\
            +  beta[6]*scaledDF['scaled_total_riders']

    # Likelihood (sampling distribution) of observations
    Y_obs = pm.Normal('Y_obs', mu=mu, sd=sigma, observed=scaledDF["nextDay"])
```

```python
map_estimate = pm.find_MAP(model=big_model)
map_estimate
```

```python
map_estimate = pm.find_MAP(model=big_model, method='powell')
map_estimate
```
